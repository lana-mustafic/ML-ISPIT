```python
import pandas as pd
import numpy as np
df=pd.read_csv('train.csv') #ucitavanje podataka
```

```python
df.head() #prvih 5 redova
df.info() #tipovi podataka
df.isnull().sum() #broj NaN vrijednosti po koloni
df.describe() #statistika za numericke kolone
```

```python
#izbacujemo kolone
#prvo ispisujem sve kolone
print(df.columns.tolist())
```

```python
#izbacujem Unnamed: 0, id i satisfaction
df=df.drop(['Unnamed: 0','id','satisfaction'],axis=1)
```

```python
#provjeravam jesu li kolone izbacene
print(df.columns.tolist())
```

```python
#provjeravam koje kolone imaju NaN i koliko?
nan_before=df.isnull().sum()
print(nan_before[nan_before>0])
```

```python
#odvajamo numericke i kategoricke kolone
numeric_cols=df.select_dtypes(include=['int64','float64']).columns
categorical_cols=df.select_dtypes(include=['object']).columns

print(f"Numericke kolone: {len(numeric_cols)}")
print(f"Kategoricke kolone: {len(categorical_cols)}")
```

```python
#popunjavamo numericke kolone sa medijanom
for col in numeric_cols:
  median_value=df[col].median()
  df[col]=df[col].fillna(median_value)
  print(f"{col}: popunjeno sa medijanom = {median_value}")
```

```python
#popunjavamo kategoricke kolone sa modom
for col in categorical_cols:
  mode_value=df[col].mode()[0]
  df[col]=df[col].fillna(mode_value)
  print(f"{col}: popunjeno sa modom = '{mode_value}'")
```

```python
#provjeravamo - treba da bude 0
nan_after=df.isnull().sum().sum()
print(f"NaN nakon popunjavanja: {nan_after}")
```

```python
#pretvaramo kategoricke kolone u numericke
from sklearn.preprocessing import LabelEncoder

df['Customer Type']=df['Customer Type'].str.capitalize()
#1. TARGET - Customer Type (pretvaramo u 0/1)
le=LabelEncoder()
df['Customer Type_encoded']=le.fit_transform(df['Customer Type'])

print("Encoding za Customer Type: ")
print(f"Klase: {le.classes_}")
print(f"0 = {le.classes_[0]},1 = {le.classes_[1]}")
```

```python
print(df.select_dtypes(include=['object']).columns.tolist()) #provjeravam koje jos kategoricke klase imaju
```

```python
#one hot encoding za preostale kategoricke klase
df=pd.get_dummies(df,columns=['Gender','Type of Travel','Class'],drop_first=True)
```

```python
#izbacujem originalnu kolonu Customer Type
df=df.drop('Customer Type', axis=1)
```

```python
#provjeravam koje kolone su ostale
print(df.columns.tolist())
```

```python
#odvajamo ulaz(x) i y(target)
x=df.drop('Customer Type_encoded',axis=1) #sve osim target
y=df['Customer Type_encoded']

print(f"Oblik x(ulazne kolone): {x.shape}")
print(f"Oblik y(target): {y.shape}")
print(f"Distribucija targeta: ")
print(y.value_counts())
print(y.value_counts(normalize=True))
```

```python
#dijelimo na train i test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    test_size=0.2, #20% za testiranje
    random_state=42, #da mozemo ponoviti rezultat
    stratify=y #zadrzava isti odnos zadovoljnih/nezadovoljnih
)
```

```python
#provjeravamo train i test
print("Podjela podataka:")
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

print("Distribucija u train: ")
print(y_train.value_counts(normalize=True))
print("Distribucija u test: ")
print(y_test.value_counts(normalize=True))

print("Provjera stratify-a")
print(f"Originalni udio klase: {y.mean():.4f}")
print(f"Udio train klase: {y_train.mean():.4f}")
print(f"Udio test klase: {y_test.mean():.4f}")
```

```python
#biranje 2 modela
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

#MODEL 1: Linear Regression (jednostavan i brz):
model1=LogisticRegression(max_iter=1000,random_state=42)

#MODEL 2: RandomForest(snazniji, ali sporiji)
model2=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

print("Podaci su skalirani za LogisticRegression!")

#treniranje modela
#TRENIRANJE 1. MODELA (Logistic Regression)
model1.fit(x_train_scaled,y_train)
y_pred1=model1.predict(x_test_scaled)

#TRENIRANJE 2. MODELA(Random Forest)
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)

print("Modeli su istrenirani!")
```

```python
#metrike
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
#logistic regression (sa skaliranjem)
print(f"Accuracy: {accuracy_score(y_test,y_pred1):.4f}")
print(f"Precision: {precision_score(y_test,y_pred1):.4f}")
print(f"Recall: {recall_score(y_test,y_pred1):.4f}")
print(f"F1: {f1_score(y_test,y_pred1):.4f}")
print(f"Confusion_matrix: ")
print(confusion_matrix(y_test,y_pred1))

#random forest(bez skaliranja)
print(f"Accuracy: {accuracy_score(y_test,y_pred2):.4f}")
print(f"Precision: {precision_score(y_test,y_pred2):.4f}")
print(f"Recall: {recall_score(y_test,y_pred2):.4f}")
print(f"F1: {f1_score(y_test,y_pred2):.4f}")
print(f"Confusion_matrix: ")
print(confusion_matrix(y_test,y_pred2))
```
